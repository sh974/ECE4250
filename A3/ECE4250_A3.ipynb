# 1. Compressed Sensing

## Part 1

```python
def RandomMask(x, f):
    m = np.ones(x.shape)
    r,c = x.shape
    
    for i in range(int(r * c * f / 100)):
        temp_x, temp_y = [int(np.random.rand() * c), int(np.random.rand() * r)]
        
        while(m[temp_x, temp_y] == 0):
            temp_x, temp_y = [int(np.random.rand() * c), int(np.random.rand() * r)]
        m[temp_x, temp_y] = 0
    
    return np.multiply(x, m), m
```

```python
x = mpimg.imread('./X.png')
y = mpimg.imread('./Y.png')
xtilde, m = RandomMask(x, 10)
plt.subplot(1,2,1)
plt.imshow(xtilde)
plt.title('Masked Image')
plt.subplot(1,2,2)
plt.imshow(m)
plt.title('Mask')
```

   ## Part 2

```python
def ImageGrad(x):
    sx = ndimage.sobel(x, 0)
    # Get y-gradient in "sy"
    sy = ndimage.sobel(x, 1)
    # Get square root of sum of squares
    sobel=np.hypot(sx,sy)

    # Hopefully see some edges
    return sobel
```

```python
sobel = ImageGrad(x)  
plt.imshow(sobel)

# verify images are the same size
print (len(sobel[0]), len(sobel[1]))
print (len(x[0]), len(x[1]))
```

## Part 3

```python
def ReconLoss(xhat, xtilde, m, l):
    grad = ImageGrad(xhat)
    grad = grad.reshape((grad.shape[0]*grad.shape[1]))
    temp = l * np.linalg.norm(grad, 1)
    
    mx = np.multiply(xhat, m)
    mx = mx.reshape((mx.shape[0]*mx.shape[1]))
    xtilde = xtilde.reshape((xtilde.shape[0]*xtilde.shape[1]))
    temp1 = np.linalg.norm(mx - xtilde, 2)**2
        
    return temp + temp1
```

```python
ReconLoss(x, xtilde, m, 5)
```

```python
def ImageMerge(x1, x2, alpha):
    return np.multiply(x1, alpha) + np.multiply(x2, 1 - alpha)
```

# 2. FIR Filter Design

## Part 1


```python
def fir_ls(N, ws, As):
    As = np.transpose(As)
    M = int((N-1)/2)
    
    # cosine matrix
    F = np.zeros((len(ws), M + 1))
    for i in range(len(ws)):
        for j in range(M+1):
            if(j == 0):
                F[i][j] = 1
            else:
                F[i][j] = 2 * np.cos(ws[i]*j)
    
    # compute pseudo-inverse
    Fi = np.linalg.pinv(F)
    
    h_temp = np.matmul(Fi, As)

    # plot on both sides
    h = np.zeros(N)
    h[0:M+1] = np.flip(h_temp)
    h[M:N] = h_temp
    return h
```

## Part 2


```python
L = 1000
ws = np.linspace(0, np.pi, L)
As = np.zeros(L)

for i in range(L):
    if (ws[i ]< np.pi / 8):
        As[i] = 1
    elif (ws[i] < np.pi / 4):
        As[i] = 2
    elif (ws[i] < 3 * np.pi / 8):
        As[i] = 1
    elif (ws[i] < np.pi / 2):
        As[i] = 0.5


h = fir_ls(101, ws, As)
plt.plot(h)
plt.title('h in Time Domain')
plt.xlabel('n')
plt.ylabel('h')
plt.show()
```

```python
H = np.fft.fft(h, L)
plt.plot(np.linspace(0, np.pi, int(L/2)), np.abs(H)[0:int(L/2)], label='My Result')
plt.plot(ws, As, label='Ideal Filter')
plt.title('H in Frequency Domain')
plt.xlabel('omega')
plt.ylabel('H')
plt.legend()
plt.show()
```

## Part 3

```python
N = 101
L = 1000
cutoff = np.pi/16
ws = np.linspace(0, np.pi, L)
As = np.zeros(1000)
for i in range(len(ws)):
    if ws[i] <= cutoff: 
        As[i] = 1
        
h = fir_ls(N, ws, As)
plt.plot(h)
plt.title('LPF with N = 101, L = 1000, in Time Domain')
plt.xlabel('n')
plt.ylabel('H')
plt.legend()
plt.show()
```

```python
H = np.fft.fft(h, L)
plt.plot(np.linspace(0, np.pi, int(L/2)), np.abs(H)[0:int(L/2)], label='My Result')
plt.plot(ws, As, label='Ideal LP Filter')
plt.title('LPF with N = 101, L = 1000, in Frequency Domain')
plt.xlabel('omega')
plt.ylabel('H')
plt.legend()
plt.show()
```
