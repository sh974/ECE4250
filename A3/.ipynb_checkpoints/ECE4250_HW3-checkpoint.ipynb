```python
import numpy as np
import imageio
from scipy import ndimage
from scipy import signal
from scipy.io import wavfile
from matplotlib import image
import matplotlib.pyplot as plt
```

## 1-1 RandomMask ##

```python
def RandomMask(x, f):
    m = np.ones(x.shape)
    nrows = x.shape[0]
    ncols = x.shape[1]
    
    for i in range(int(nrows*ncols*f/100)):
        px = int(np.random.rand()*ncols)
        py = int(np.random.rand()*nrows)
        while(m[px, py] == 0):
            px = int(np.random.rand()*ncols)
            py = int(np.random.rand()*nrows)
        m[px, py] = 0
    
    x_tilt = np.multiply(x, m)
    return x_tilt, m
```

```python
x = np.asarray(Image.open('X.png')).astype('float64')
y = np.asarray(Image.open('Y.png')).astype('float64')

(xtilt, m) = RandomMask(x, 10)
imageio.imsave('xtilt.png', xtilt)
imageio.imsave('m.png', m)
plt.subplot(1,2,1)
plt.imshow(xtilt)
plt.title('Masked Image X')
plt.subplot(1,2,2)
plt.imshow(m)
plt.title('Mask of X')
plt.show()
```

## Q1-2 ImageGrad ##

```python
def ImageGrad(x):
    gradx = ndimage.sobel(x, 0)
    grady = ndimage.sobel(x, 1)
    return np.sqrt(np.power(gradx, 2) + np.power(grady, 2))
```

```python
grad = ImageGrad(x)
plt.imshow(grad.astype(float))
plt.title('Sobel Gradient of X')
plt.show()
```

## Q1-3 ReconLoss ##

```python
def ReconLoss(xtilt, x, m, lamb):
    mx = np.multiply(x, m)
    gradx = ImageGrad(x)
    # to avoid messing up with the norm function reshape everything to 1d
    xtilt_vec = xtilt.reshape((xtilt.shape[0]*xtilt.shape[1]))
    mx_vec = mx.reshape((mx.shape[0]*mx.shape[1]))
    gradx_vec = gradx.reshape((gradx.shape[0]*gradx.shape[1]))
    return np.linalg.norm(mx_vec-xtilt_vec, ord=2)**2 + lamb*np.linalg.norm(gradx_vec, ord=1)
```

## Q1-4 ImageMerge ##

```python
def ImageMerge(x1, x2, a):
    return np.multiply(x1, a) + np.multiply(x2, 1-a)
```

```python
plt.imshow(ImageMerge(x, y, 0.5))
plt.title('Merge of 50% x and 50% y')
plt.show()
```

## Q1-5 ImageMergeWithMask ##

```python
def ImageMergeWithMask(x1, x2, m, a):
    return np.multiply(m, x1) + np.multiply(1-m, ImageMerge(x1, x2, a))
```

```python
plt.imshow(ImageMergeWithMask(x, y, m, 1))
plt.title('Merge of 50% x and 50% y with Mask')
plt.show()
```

## Q1-6 Reconstruct Loss Without Mask ##

```python
loss_list = np.zeros((100,))
# a from 0 to 1 by 0.01
for i in range(100):
    x_recon = ImageMerge(x, y, i/100)
    loss_list[i] = ReconLoss(xtilt, x_recon, m, 20)
    
plt.plot(np.arange(0, 1, 0.01), loss_list)
plt.title('Reconstruction Loss of ImageMerge')
plt.xlabel('a')
plt.ylabel('loss')
plt.show()
```

```python
print('The optimal a is: ', np.argmin(loss_list)/100, ", where the loss is: ", np.amin(loss_list)/1e8, "*1e8")
```

## Q1-7 Reconstruct Loss With Mask ##

```python
masked_loss_list = np.zeros((100,))
# a from 0 to 1 by 0.01
for i in range(100):
    x_recon = ImageMergeWithMask(x, y, m, i/100)
    masked_loss_list[i] = ReconLoss(xtilt, x_recon, m, 20)
    
plt.plot(np.arange(0, 1, 0.01), masked_loss_list)
plt.title('Reconstruction Loss of ImageMergeWithMask')
plt.xlabel('a')
plt.ylabel('loss')
plt.show()
```

```python
print('The optimal a is: ', np.argmin(masked_loss_list)/100, ", where the loss is: ", np.amin(masked_loss_list)/1e8, "*1e8")
```

## Q1-8 The Optimal Reconstruction ##

```python
plt.subplot(1,3,1)
plt.imshow(x)
plt.title('The Actual Image')
plt.subplot(1,3,2)
plt.imshow(ImageMerge(x, y, 0.8))
plt.title('ImageMerge with a=0.8')
plt.subplot(1,3,3)
plt.imshow(ImageMergeWithMask(x, y, m, 0.93))
plt.title('ImageMergeWithMask with a=0.93')

plt.subplots_adjust(left=0.1, right=2.1, bottom = 0.1, top = 2.1, wspace=0.4, hspace=0.4)
plt.show()
```

I am expecting the LS error of ImageMergeWithMask to be lower because it utilizes the information from the mask, but it is actually higher. This is probably because of the regularizaiton. If I turn of the regularization the error of ImageMergeWithMask gets significantly lower. <br /><br />

ImageMergeWithMask gives a higher a, which means that it is more confident that the original image is a racoon (it that a racoon?) not a squirrel. Again the reasone is that ImageMergeWithMask already knows what's in the unmasked region so it is more confident. <br /><br />

To me it seems that the three are almost the same. However the one with mask captures more details than the one without the mask, like the texture on the leaf and the racoon's face. This makes sense because ImageMergeWithMask only applies changes to the masked pixels while ImageMerge works on the entire image.

## Q2 FIR Filter Design ##

## 2-1 ##

```python
def genT1FIR(N, ws, As):
    M = (int)((N-1)/2)
    A = np.zeros((len(ws), M+1))
    for i in range(len(ws)):
        for j in range(M+1):
            if(j == 0):
                A[i][j] = 1
            else:
                A[i][j] = 2*np.cos(ws[i]*j)
    Ainv = np.linalg.pinv(A)
    h_raw = np.matmul(Ainv, np.transpose(As))
    h = np.zeros((N,))
    h[0:M+1] = np.flip(h_raw)
    h[M:N] = h_raw
    return h
```

## 2-2 ##

```python
ws = np.linspace(0, np.pi, 1000)
As = np.zeros((1000,))
for i in range(1000):
    if(ws[i]<np.pi/8):
        As[i] = 1
    elif(ws[i]<np.pi/4):
        As[i] = 2
    elif(ws[i]<np.pi*3/8):
        As[i] = 1
    elif(ws[i]<np.pi/2):
        As[i] = 0.5

h = genT1FIR(101, ws, As)
plt.plot(h)
plt.title('Time Domain Response')
plt.xlabel('n')
plt.ylabel('h')
plt.show()
```

```python
H = np.fft.fft(h, 10000)
plt.plot(np.linspace(0, np.pi, 5000), np.abs(H)[0:5000], label='my filter')
plt.plot(ws, As, label='ideal filter')
plt.title('Frequency Response')
plt.xlabel('omega')
plt.ylabel('H')
plt.legend()
plt.show()
```

## 2-3 ##

```python
ws = np.linspace(0, np.pi, 1000)
As = np.zeros((1000,))
for i in range(1000):
    if(ws[i]<np.pi/16):
        As[i] = 1

my_h = genT1FIR(101, ws, As)
plt.plot(my_h)
plt.title('Time Domain Response of my LPF')
plt.xlabel('n')
plt.ylabel('h')
plt.show()
```

```python
my_H = np.fft.fft(my_h, 10000)
plt.plot(np.linspace(0, np.pi, 5000), np.abs(my_H)[0:5000], label='my filter')
plt.plot(ws, As, label='ideal filter')
plt.title('Frequency Response of my LPF')
plt.xlabel('omega')
plt.ylabel('H')
plt.legend()
plt.show()
```

## 2-4 ##

```python
# firls by default sets the sampling freq to 2
ws = ws/np.pi
firl_h = signal.firls(101, ws, As)
plt.plot(firl_h)
plt.title('Time Domain Response of LPF from firl')
plt.xlabel('n')
plt.ylabel('h')
plt.show()
```

```python
ws = ws*np.pi
firl_H = np.fft.fft(firl_h, 10000)
plt.plot(np.linspace(0, np.pi, 5000), np.abs(firl_H)[0:5000], label='firls')
plt.plot(ws, As, label='ideal filter')
plt.title('Frequency Response of my LPF')
plt.xlabel('omega')
plt.ylabel('H')
plt.legend()
plt.show()
```

```python
plt.plot(np.linspace(0, np.pi, 5000), np.abs(my_H)[0:5000], color='red', label='my LPF')
plt.plot(np.linspace(0, np.pi, 5000), np.abs(firl_H)[0:5000], color='blue', label='firls')
plt.title('Comparison of Two LPF')
plt.legend()
plt.xlabel('omega')
plt.ylabel('H')
plt.show()
```

The Two are almost the same and it is hard two tell the two lines from the graph
